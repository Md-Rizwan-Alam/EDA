Question: 1. What is the average rating for each product category?

```python
#Answer
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("amazon.csv")
```

```python
df
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with NaN values in the 'rating' column
df = df.dropna(subset=['rating'])

# Grouping by 'category' and calculating the mean of 'rating' for each category
average_rating_per_category = df.groupby('category')['rating'].mean()

# result
print("Average rating for each product category:")
print(average_rating_per_category)
```

Question: 2. What are the top rating_count products by category?

```python
#Answer
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Drop rows with NaN values in the 'rating_count' column
df = df.dropna(subset=['rating_count'])

# Define a function to select the top products within each category
def top_products_by_category(group):
    return group.nlargest(5, 'rating_count')

# Grouping by 'category' and applying the function to select top products
top_products_by_category = df.groupby('category').apply(top_products_by_category)

# Resetting the index to flatten the DataFrame
top_products_by_category = top_products_by_category.reset_index(drop=True)

# result
print("Top rating_count products by category:")
print(top_products_by_category)
```

Question: 6.  What are the most popular product keywords?

```python
#Answer
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt



words = df['product_name'].str.split()


flat_words = [word for sublist in words for word in sublist]


word_counts = Counter(flat_words)


word_counts_df = pd.DataFrame.from_dict(word_counts, orient='index', columns=['count'])


word_counts_df = word_counts_df.sort_values(by='count', ascending=False)


top_keywords = word_counts_df.head(10)


plt.figure(figsize=(10, 6))
top_keywords.plot(kind='bar', legend=None)
plt.xlabel('Keyword')
plt.ylabel('Frequency')
plt.title('Top 10 Most Popular Product Keywords')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Question: 8. What is the correlation between discounted_price and rating?

```python
#Answer
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset

# First, we need to ensure that the 'discounted_price' column is numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Drop rows where 'discounted_price' or 'rating' is NaN
df = df.dropna(subset=['discounted_price', 'rating'])

# Calculating the correlation between 'discounted_price' and 'rating'
correlation = df['discounted_price'].corr(df['rating'])

# Plotting the correlation using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='discounted_price', y='rating')
plt.title('Correlation between Discounted Price and Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.grid(True)
plt.show()

print("Correlation between discounted_price and rating:", correlation)
```

Question: 9. What are the Top 5 categories based on the highest ratings?

```python
#Answer
```

```python
df = pd.read_csv("amazon.csv")
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset

# Clean the 'rating' column by removing unexpected string values and converting to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with missing values in 'rating' column
df = df.dropna(subset=['rating'])

# Grouping the data by category and calculating the mean rating for each category
average_rating_per_category = df.groupby('category')['rating'].mean()

# Sorting the categories based on the mean rating in descending order
sorted_categories = average_rating_per_category.sort_values(ascending=False)

# Selecting the top 5 categories with the highest mean ratings
top_5_categories = sorted_categories.head(5)

# Plotting the top 5 categories based on the highest ratings
plt.figure(figsize=(10, 6))
top_5_categories.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Top 5 Categories with Highest Ratings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Question: 1.Load the dataframe and ensure data quality by checking for missing values and duplicate rows.
 Handle missing values and remove duplicate rows if necessary.

```python
#Answer
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("spotify.csv")


plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title("Missing Values Check")
plt.show()


duplicate_rows = df.duplicated().sum()
if duplicate_rows > 0:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df.duplicated(), palette='Set2')
    plt.title("Duplicate Rows Check")
    plt.xlabel("Duplicate Rows")
    plt.ylabel("Count")
    plt.xticks([0, 1], ['Unique', 'Duplicate'])
    plt.show()


missing_column = 'Popularity'
plt.figure(figsize=(10, 6))
sns.histplot(df[missing_column], bins=20, kde=True, color='skyblue')
plt.title(f"Distribution of {missing_column}")
plt.xlabel(missing_column)
plt.ylabel("Frequency")
plt.show()


df['Popularity'].fillna(df['Popularity'].median(), inplace=True)


if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicate rows removed.")


duplicate_rows_after_removal = df.duplicated().sum()
print("Duplicate Rows after Removal:", duplicate_rows_after_removal)


print("\nCleaned DataFrame:")
print(df.head())
```

Question: 7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist

```python
#Answer
```

```python
top_5_most_popular_artists = average_popularity_by_artist.head(5).index.tolist()

top_5_artists_tracks = df[df['Artist'].isin(top_5_most_popular_artists)]

most_popular_tracks_by_artist = top_5_artists_tracks.groupby('Artist').apply(lambda x: x.loc[x['Popularity'].idxmax()])

print("Most Popular Tracks for the Top 5 Most Popular Artists:")
for artist, track in most_popular_tracks_by_artist[['Artist', 'Track Name']].iterrows():
    print(f"Artist: {track['Artist']}, Most Popular Track: {track['Track Name']}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_5_artists_tracks = df[df['Artist'].isin(top_5_most_popular_artists)]

most_popular_tracks_by_artist = top_5_artists_tracks.groupby('Artist').apply(lambda x: x.loc[x['Popularity'].idxmax()])

plt.figure(figsize=(10, 6))
sns.barplot(x='Track Name', y='Popularity', hue='Artist', data=most_popular_tracks_by_artist, palette='Set1')
plt.xlabel('Track Name')
plt.ylabel('Popularity')
plt.title('Most Popular Tracks for Top 5 Most Popular Artists')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Artist', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
